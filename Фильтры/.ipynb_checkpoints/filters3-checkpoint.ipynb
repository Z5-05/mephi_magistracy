```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft, fftshift
from scipy import signal

import warnings
warnings.filterwarnings("ignore")
```

```python
def plt_sel(s, *args, **kwargs):
    if s == 0:
        return plt.plot(*args)
    if s == 1:
        return plt.stem(*args, **kwargs)
    if s == 2:
        return plt.step(*args)
```

```python
# constants
X_MIN = 0
X_MAX = 3

# кол-во снимаемых данных
N_SAMPLES = 30
# частота дискретизации
FS = (X_MAX - X_MIN)/N_SAMPLES
```

```python
ws = 0.5
```

#### Функция сама - синусоида:

```python
# time vector
ts = np.linspace(X_MIN, X_MAX, N_SAMPLES, endpoint=True)
fi = np.pi/2

# sine wave
x = np.sin(ws*np.pi*ts + fi)
```

```python
plt_sel(1
      , ts
      , x
      , use_line_collection=True
      , linefmt="g")
plt.show()
```

```python
L = 8
new_n = L * (N_SAMPLES-1)
```

```python
new_t = np.linspace(X_MIN, X_MAX, new_n+1, endpoint=True)

new_x = np.empty(new_n+1)
new_x.fill(0)

for i, _ in enumerate(x):
    new_x[i*L] = x[i]
```

```python
plt_sel(1
      , new_t
      , new_x
      , use_line_collection=True
      , linefmt="b")
plt.show()
```

```python
# длина фильтра
n_w = int(new_n / L)

# Частота дискретизации входных данных
fd = FS
# Частота полосы пропускания
wp = 0.5 * np.pi
# Частота полосы затухания
ws = wp + 0.2 * np.pi

# Импульсная характеристика фильтра
h_low_pass = np.zeros(n_w)
# Идеальная импульсная характеристика
h = np.zeros(n_w) 
# Весовая функция значения
w = np.zeros(n_w) 

# Расчет импульсной характеристики фильтра
fc = (wp + ws) / 2
```

```python
for i in range(-int(n_w / 2), int(n_w / 2)):
    h_low_pass[i + int(n_w / 2)] = np.sinc(fc * i / np.pi)
```

```python
for i in range(n_w):
    # весовые коэффициенты окна Хэмминга
    w[i] = 0.53863 - 0.46164 * math.cos((2 * math.pi * i) / ( n_w-1))
    #w[i] = 1
```

```python
# for i in range(n_w):
#     # весовые коэффициенты окна Хэмминга
#     h[i] = h_low_pass[i] * w[i]
```

```python
h = np.convolve(h_low_pass, w, mode='same')
```

```python
plt.plot(h_low_pass)
plt.plot(w)
plt.plot(h)
plt.grid()
plt.show()
```

```python
new_x2 = np.convolve(new_x, h, mode='full')
```

```python
new_n2 = len(new_x2)
tt2 = np.linspace(X_MIN, X_MAX, new_n2, endpoint=True)
```

```python
plt_sel(1, tt2, new_x2, linefmt='b')
plt_sel(1, ts, x, linefmt='r')
plt.show()
```
